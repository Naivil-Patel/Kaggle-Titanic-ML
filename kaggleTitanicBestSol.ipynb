```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
train = pd.read_csv("train.csv")
train['Embarked'].fillna(3, inplace=True)
```

```python
train.dropna()
sex = {'male': 1,'female': 0}
train.Sex = [sex[item] for item in train.Sex]
train['Embarked'] = train['Embarked'].replace(['S','C','Q'],['0','1','2'])
```

```python
y = train["Survived"]
X = train.drop(['PassengerId','Survived','Ticket','Name','Cabin'],axis = 1)
```

```python
X=X.fillna("29.699118")
```

```python
X
```

```python
clf = LogisticRegression().fit(X,y)
clf2 = RandomForestClassifier(max_depth = 7, random_state = 0) #ideal is 7. More than 7 will overfit the training set.
clf2.fit(X,y)
```

```python
clf.score(X,y)
clf2.score(X,y) #Random forrest has always done better than Logistic Regression
```

```python
test=pd.read_csv("test.csv")
sex = {'male': 1,'female': 0}
test.Sex = [sex[item] for item in test.Sex]
test['Embarked'].fillna(3, inplace=True)
test['Embarked'] = test['Embarked'].replace(['S','C','Q'],['0','1','2'])
test.describe()
```

```python
X_test = test.drop(['PassengerId','Ticket','Name','Cabin'],axis = 1)
X_test=X_test.fillna("30.272590")
test
```

```python
y=clf.predict(X_test)
y2=clf2.predict(X_test)
```

```python
df = pd.DataFrame({'PassengerId':test["PassengerId"], 'Survived':y})
df2 = pd.DataFrame({'PassengerId':test["PassengerId"], 'Survived':y2})
```
